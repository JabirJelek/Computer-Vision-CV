Create the Directory Structure:

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Define paths
drive_path = '/content/drive/MyDrive/Object Detection/data'
labels_source = os.path.join(drive_path, 'newYoloLabel') # Change this, use the converted label in the different directory
images_source = os.path.join(drive_path, 'new_data/frames')
dataset_path = '/content/drive/MyDrive/Object Detection/data/dataset-new'  # New structured dataset location

# Create main dataset directories
os.makedirs(os.path.join(dataset_path, 'images', 'train'), exist_ok=True)
os.makedirs(os.path.join(dataset_path, 'images', 'val'), exist_ok=True)
os.makedirs(os.path.join(dataset_path, 'labels', 'train'), exist_ok=True)
os.makedirs(os.path.join(dataset_path, 'labels', 'val'), exist_ok=True)
```

Get and Split File List:

```python
# Get all image filenames (without extensions)
image_files = [os.path.splitext(f)[0] for f in os.listdir(images_source)
               if f.endswith(('.jpg', '.jpeg', '.png'))]

# Split into train/val (80% train, 20% val)
train_files, val_files = train_test_split(image_files, test_size=0.2, random_state=42)
```

Copy Files to Structured Folders:

```python
# Helper function to copy files
def copy_files(files, source_img, source_lbl, dest_img, dest_lbl):
    for file in files:
        # Copy image
        for ext in ['.jpg', '.jpeg', '.png']:
            src_img_path = os.path.join(source_img, file + ext)
            if os.path.exists(src_img_path):
                shutil.copy(src_img_path, os.path.join(dest_img, file + ext))
                break

        # Copy label
        src_lbl_path = os.path.join(source_lbl, file + '.txt')
        if os.path.exists(src_lbl_path):
            shutil.copy(src_lbl_path, os.path.join(dest_lbl, file + '.txt'))

# Copy training files
copy_files(train_files, images_source, labels_source,
           os.path.join(dataset_path, 'images', 'train'),
           os.path.join(dataset_path, 'labels', 'train'))

# Copy validation files
copy_files(val_files, images_source, labels_source,
           os.path.join(dataset_path, 'images', 'val'),
           os.path.join(dataset_path, 'labels', 'val'))
```
