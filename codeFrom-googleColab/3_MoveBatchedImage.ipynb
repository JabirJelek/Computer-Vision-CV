```python
import os
import shutil

def move_txt_files_only(base_path, destination_path):
    """
    Moves all TXT files from base_path to destination_path.
    No assumptions about image files or other counterparts.

    Args:
        base_path (str): Directory containing files to process
        destination_path (str): Target directory for moving TXT files
    """
    # Ensure destination directory exists
    os.makedirs(destination_path, exist_ok=True)

    # Get all files in base directory
    try:
        all_files = os.listdir(base_path)
        print(f"Found {len(all_files)} files in the directory.")
    except FileNotFoundError:
        print(f"Error: The directory '{base_path}' does not exist. Please check the path.")
        return

    # Find all TXT files
    txt_files = [file for file in all_files if file.endswith('.txt')]
    print(f"Found {len(txt_files)} TXT files.")

    moved_count = 0

    # Move all TXT files
    for txt_file in txt_files:
        source_path = os.path.join(base_path, txt_file)
        destination_file_path = os.path.join(destination_path, txt_file)

        # Move the file
        shutil.move(source_path, destination_file_path)
        print(f"Moved: {txt_file}")
        moved_count += 1

    print(f"\nOperation complete! Successfully moved {moved_count} TXT files to '{destination_path}'.")

# --- CONFIGURATION ---
base_path = '/content/drive/MyDrive/ObjectDetection-Tasks/APD-Detection/dataset0/image'
destination_path = '/content/drive/MyDrive/ObjectDetection-Tasks/APD-Detection/dataset0/textBatch-238'

# --- EXECUTE ---
move_txt_files_only(base_path, destination_path)
```
